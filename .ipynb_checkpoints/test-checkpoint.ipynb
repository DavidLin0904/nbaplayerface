```python
# import
from pymongo import MongoClient
from bson.objectid import ObjectId
from nba_api.stats.static import players
from nba_api.stats.endpoints import playergamelog

# connection
conn = MongoClient("mongodb://localhost")
db = conn.myNBA
collection = db.player

# playerid = [str(player["id"]) for player in players.get_active_players()]
ids = ["1630173", "203500"]
for i in range(len(ids)):
    playergamelogs = playergamelog.PlayerGameLog(ids[i]).get_dict()['resultSets'][0]
    headers = playergamelogs['headers']
    data = playergamelogs['rowSet']
    games = [game for game in data]
    
    for i in range(len(games)):
        for j in range(len(headers)):
            myplayer = {headers[j]:games[i][j]}
            updateplayer = {"$set": {headers[j]:games[i][j]}}
            collection.update_one(myplayer, updateplayer, True)
#             print(tt.raw_result)
#             if tt.raw_result['updatedExisting'] == True:
#                 collection.insert_one(myplayer)
cursor = collection.find()
data = [d for d in cursor] 
print(data)
```
