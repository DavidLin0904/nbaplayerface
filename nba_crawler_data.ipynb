```python
def update_insert(headers, games, updateset=True):
    for i in range(len(headers)):
        for j in range(len(games)):
            if len(games) != 0: # 出賽場數
                myplayer = {headers[i]:games[j][i]}
                updateplayer = {"$set": {headers[i]:games[j][i]}}
                collection.update_one(myplayer, updateplayer, True)
```

```python
# import
from pymongo import MongoClient
from bson.objectid import ObjectId
from nba_api.stats.static import players
from nba_api.stats.endpoints import playergamelog
import time
# connection
conn = MongoClient("mongodb://localhost")
db = conn.NBAPlayerData
collection = db.playerData
playerid = [str(player["id"]) for player in players.get_active_players()]
for ids in playerid:
    playergamelogs = playergamelog.PlayerGameLog(ids).get_dict()['resultSets'][0]
    headers = playergamelogs['headers']
    data = playergamelogs['rowSet']
    games = [game for game in data]
    update_insert(headers, games)
        
cursor = collection.find()
data = [d for d in cursor] 
print(data)
```
